```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/_titanic.csv')
```

# HISTOGRAM

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=30,color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram')
```

BAR CHART

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Pclass'], y=df['Age'], estimator=sum, errorbar=None)
plt.title('Total Age by Passenger Class')
plt.show()
```

BOX PLOT

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Pclass'], y=df['Age'])
plt.title('Box Plot of Age by Passenger Class')
plt.show()
```

PAIR PLOT

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass']])
plt.show()
```

CORRECTION  HEATMAP

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Age', 'Fare', 'Pclass']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

VIOLIN PLOT

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['Pclass'], y=df['Age'])
plt.title('Violin Plot of Age by Passenger Class')
plt.show()
```

COUNT PLOT

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Pclass'])
plt.title('Count of Passengers by Class')
plt.show()
```

LINE CHART

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(len(df)), y=df['Age'])
plt.title('Age Distribution Over Index')
plt.show()
```
